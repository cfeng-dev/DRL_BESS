```python
import matplotlib.pyplot as plt
from utils.csv_handler import load_price_data
```

```python
csv_path = "../../data/electricity_price/dayahead_2024_11.csv"

df_p, price_series, price_ts = load_price_data(
    csv_path,
    resolution="15min",
)

df_p.head()
```

```python
plt.figure(figsize=(18, 6))

plt.plot(
    df_p["Day Ahead Auktion"],
    color="purple",
    label="Day-Ahead Electricity Price"
)

plt.title("Electricity Price", fontsize=18, fontweight="bold")
plt.ylabel("Price [EUR/MWh]", fontsize=16)
plt.xlabel("Time", fontsize=16)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.legend(loc="upper right", fontsize=13, frameon=True)
plt.grid(True)
plt.tight_layout()
plt.show()
```
